```python
from transformers import pipeline
```

```python
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """
The James Webb Space Telescope has captured stunning new images of the Carina Nebula, a stellar nursery where stars are born.
The images reveal intricate details of the nebula's gas and dust clouds, providing scientists with valuable insights into the process of star formation.
The telescope's powerful infrared cameras have allowed astronomers to peer through the obscuring dust and see newborn stars in unprecedented detail.
These observations are helping researchers understand the early stages of stellar evolution and the formation of planetary systems.
"""
```

```python
summary = summarizer(text, max_length=150, min_length=30, do_sample=False)
```

```python
print(summary[0]['summary_text'])
```
